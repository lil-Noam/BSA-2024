### Question 15
A company's product quality control process involves two independent tests. Each test has a 98% probability of detecting a defect if it is present.
1. What is the probability that both tests detect a defect when it is present?
2. What is the probability that at least one test detects a defect when it is present?
3. If the company requires both tests to detect a defect before rejecting a product, calculate the probability that a defective product will be rejected.

$ D $ = a defect is present <br>
$ \neg D $ = a defect is not present <br>
$ T_1^+ $ = first test detects defect <br>
$ T_1^- $ = first test does not detect defect <br>
$ T_2^+ $ = second test detects defect <br>
$ T_2^- $ = second test does not detect defect <br>

  probabilities: <br>
$ P(T_1^+ | D) = 0.98 $ <br>
$ P(T_1^- | D) = 1-0.98 = 0.02 $ <br>
$ P(T_2^+ | D) = 0.98 $ <br>
$ P(T_2^- | D) = 1-0.98 = 0.02 $ <br>
<br><br>

1. the probability that both tests detect a defect when it is present: <br>
$$ P(T_1^+ \land T_2^+ | D) $$
since the test are independent: <br>
$$ P(T_1^+ \land T_2^+ | D) = P(T_1^+ | D) \cdot P(T_2^+ | D) = 0.98 \cdot 0.98 = 0.9604 $$ <br><br>

2. "at lest one test detects a defect when it is present" is the complement of "no test detects a defect when it is present": <br>
$$ P(T_1^+ \lor T_2^+ | D) = 1 - P(T_1^- \land T_2^- | D) $$ <br>
since the test are independent: <br>
$$  P(T_1^- \land T_2^- | D) = P(T_1^- | D) \cdot P(T_2^- | D) = 0.02 \cdot 0.02 = 0.0004 $$ <br>
so: <br>
$$ P(T_1^+ \lor T_2^+ | D) = 1 - 0.0004 = 0.9996 $$ <br><br>

3. If the company requires both tests to detect a defect before rejecting a product, the probability that a defective product will be rejected is the same probability that both tests will detect the defect (calculated in 1) <br>
$$ P(T_1^+ \land T_2^+ | D) = P(T_1^+ | D) \cdot P(T_2^+ | D) = 0.98 \cdot 0.98 = 0.9604 $$ <br><br>
